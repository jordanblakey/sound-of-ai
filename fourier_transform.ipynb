```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
audio_path = "audio/piano_c.wav"
ipd.Audio(audio_path)
```

```python
signal, sr = librosa.load(audio_path)
```

```python
plt.figure(figsize=(18, 8))
librosa.display.waveshow(signal, sr=sr, alpha=0.5)
plt.show()
```

```python
ft = sp.fft.fft(signal)
magnitude = np.absolute(ft)
frequency = np.linspace(0, sr, len(magnitude))
```

```python
plt.figure(figsize=(18, 8))
plt.plot(frequency[:5000], magnitude[:5000]) # magnitude spectrum
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```

```python
# zoom in to the waveform
samples = range(len(signal))
t = librosa.samples_to_time(samples, sr = sr)

plt.figure(figsize = (18, 8))
plt.plot(t[10000:10400], signal[10000:10400])
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Create 2 cancelling sinusoids

f = 523
phase = 0
phase2 = 0.5

sin = 0.5 * np.sin(2 * np.pi * (f * t - phase))
sin2 = 0.5 * np.sin(2 * np.pi * (f * t - phase2))

sin3 = sin + sin2

plt.figure(figsize = (18, 8))
plt.plot(t[10000:10100], sin3[10000:10100], color="r")

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.ylim(-6, 6)
plt.show()
```

```python
# Create a sinusoid

f = 523
phase = 0
phase2 = 0.25
phase3 = 0.5

sin = 0.5 * np.sin(2 * np.pi * (f * t - phase))
sin2 = 0.5 * np.sin(2 * np.pi * (f * t - phase2))
sin3 = 0.5 * np.sin(2 * np.pi * (f * t - phase3))

plt.figure(figsize = (18, 8))
plt.plot(t[10000:10400], sin[10000:10400], color="r")
plt.plot(t[10000:10400], sin2[10000:10400], color="y")
plt.plot(t[10000:10400], sin3[10000:10400], color="b")

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Create a sinusoid

f = 523
phase = 0.55

sin = 0.25 * np.sin(2 * np.pi * (f * t - phase))

plt.figure(figsize = (18, 8))
plt.plot(t[10000:10400], signal[10000:10400])
plt.plot(t[10000:10400], sin[10000:10400], color="r")

plt.fill_between(t[10000:10400], sin[10000:10400] * signal[10000:10400], color="y")

plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```
